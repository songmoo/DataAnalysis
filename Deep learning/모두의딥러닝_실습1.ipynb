```python
import tensorflow as tf
import pandas as pd
```

```python
sess = tf.Session()
```

```python
train_df = pd.read_csv("data/Auction_master_kr/Auction_master_train.csv")
test_df = pd.read_csv("data/Auction_master_kr/Auction_master_test.csv")
```

```python
train_df.columns
```

```python
train_df.info()
```

```python
x1_data = train_df['Minimum_sales_price']
x2_data = train_df['Claim_price']
x3_data = train_df['Total_appraisal_price']
y_data = train_df['Hammer_price']
```

```python
x1=tf.placeholder(tf.float32)
x2=tf.placeholder(tf.float32)
x3=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
```

```python
w1=tf.Variable(tf.random_normal([1]),name='Minimum_sales_price')
w2=tf.Variable(tf.random_normal([1]),name='Total_land_real_area')
w3=tf.Variable(tf.random_normal([1]),name='Total_appraisal_price')
b=tf.Variable(tf.random_normal([1]),name='Hammer_price')
```

```python
hypothesis = x1*w1+x2*w2+x3*w3+b
cost = tf.reduce_mean(tf.square(hypothesis - y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-19)
train=optimizer.minimize(cost)
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for i in range(2001):
    cost_val, hy_val, _ = sess.run([cost,hypothesis,train],
        feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, y: y_data})
    if i % 100 == 0:
        print(i,cost_val,hy_val)
```

```python
print("예측값 : ",sess.run(hypothesis, feed_dict={x1: x1_data, x2: x2_data, x3: x3_data}))
print("실제값 : ",sess.run(y, feed_dict={y: y_data}))
```

```python
import numpy as np
def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    # noise term prevents the zero division
    return numerator / (denominator + 1e-7)
```

```python
x1_data = MinMaxScaler(train_df['Minimum_sales_price'])
x2_data = MinMaxScaler(train_df['Claim_price'])
x3_data = MinMaxScaler(train_df['Total_appraisal_price'])
```

```python
hypothesis = x1*w1+x2*w2+x3*w3+b
cost = tf.reduce_mean(tf.square(hypothesis - y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-2)
train=optimizer.minimize(cost)
sess.run(tf.global_variables_initializer())
for i in range(2001):
    cost_val, hy_val, _ = sess.run([cost,hypothesis,train],
        feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, y: y_data})
    if i % 100 == 0:
        print(i,cost_val,hy_val)
```

```python
print("예측값 : ",sess.run(hypothesis, feed_dict={x1: x1_data, x2: x2_data, x3: x3_data}))
print("실제값 : ",sess.run(y, feed_dict={y: y_data}))
```

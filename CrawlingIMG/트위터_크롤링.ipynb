```python
from selenium.webdriver.common.keys import Keys
import time
elem = driver.find_element_by_name("q")
elem.clear()
no_of_pagedowns = 1
a=0
while no_of_pagedowns:
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
    a +=1
    
    if a == 200:
        no_of_pagedowns = 0
```

```python
words=[]
for texts in driver.find_elements_by_class_name("js-tweet-text-container"):
    words.append(texts.text)
```

```python
words
```

```python
from konlpy.tag import Kkma
```

```python
kkma = Kkma()
```

```python
re
for i in words:
    re.append(kkma.nouns(i))
len(re)
```

```python
sum_re=sum(re, [])
sum_re
```

```python
nouns = [n for n in sum_re if len(n) > 1]
nouns
```

```python
from collections import Counter
```

```python
#count = Counter(nouns)
ignore=['날씨','이번','12','주신']
count = Counter(x for x in nouns if x not in ignore)
tags = count.most_common(100)
tags
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
import matplotlib.pyplot as plt
import platform
```

```python
from matplotlib import font_manager, rc
path = "C:/Windows/Fonts/malgun.ttf"
if platform.system() == "Windows":
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system()=="Darwin":
    rc('font', family='AppleGothic')
else:
    print("Unknown System")
```

```python
wc = WordCloud(font_path=path, background_color="white", max_words=2000,
               contour_width=3, contour_color='steelblue')
```

```python
#wc = WordCloud(font_path=path, background_color='white', width=800, height=600)
cloud = wc.generate_from_frequencies(dict(tags))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
```

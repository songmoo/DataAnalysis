```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import os
import time
os.getcwd()
```

```python
driver = webdriver.Chrome('chromedriver_win32/chromedriver')
driver.get('https://google.com')
```

```python
keyword = input("검색할 이미지를 입력하세요 : ")
url = 'https://www.google.co.kr/search?q='+keyword+'&source=lnms&tbm=isch&sa=X&ved=0ahUKEwic-taB9IXVAhWDHpQKHXOjC14Q_AUIBigB&biw=1842&bih=990'
```

```python
if not os.path.isdir('data/pic/'+keyword):
    os.mkdir('data/pic/'+keyword)
```

```python
driver.get(url)
```

```python
from selenium.webdriver.common.keys import Keys
elem = driver.find_element_by_name("q")
elem.clear()
no_of_pagedowns = 1
a=0
while no_of_pagedowns:
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
    a +=1
    if a == 30:
        driver.find_element_by_xpath('//*[@id="smb"]').click()
    
    if a == 100:
        no_of_pagedowns = 0
        
```

```python
import json
import urllib.request
i=0

html = driver.page_source
soup = BeautifulSoup(html, 'lxml')

for div in soup.find_all('div', class_= 'rg_meta notranslate'):
#for div in driver.find_elements_by_xpath('//*[@id="rg_s"]/div[1]/div/text()'):
    try:
        data = json.loads(div.text)
    except Exception:
        continue
    if 'ou' in data:
        print(data['ou'])
        name = "data/pic/{}/{}.{}.jpg".format(keyword,keyword,i)
        try:
            urllib.request.urlretrieve(data['ou'], name)
        except urllib.request.HTTPError:
            continue
        except urllib.request.UnicodeEncodeError:
            continue
        i +=1
        time.sleep(1)
        
        if i > 200:
            print('done')
            stop
```

```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import os
import time
os.getcwd()
```

```python
driver = webdriver.Chrome('chromedriver_win32/chromedriver')
driver.get('https://store.steampowered.com/games/')
```

```python
driver.find_element_by_xpath('//*[@id="tab_select_TopSellers"]').click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
```

```python
soup.find('div', {'id' : 'TopSellersTable'}).find_all('div', {'class' : 'tab_item_name'})
```

```python
driver.find_element_by_xpath('//*[@id="TopSellers_btn_next"]').click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
soup.find('div', {'id' : 'TopSellersRows'}).find_all('div', {'class' : 'tab_item_name'})
```

```python
driver.find_element_by_xpath('//*[@id="TopSellers_links"]/span[1]').click()
```

```python
topselling=[]
genre_top = []
for n in range (9) :
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    ranks = soup.find('div', {'id' : 'TopSellersTable'}).find_all('div', {'class' : 'tab_item_name'})
    tags =  soup.find('div', {'id' : 'TopSellersTable'}).find_all('span', {'class' : 'top_tag'})
    for rank in ranks :
        topselling.append(rank.get_text().strip())
    for tag in tags :
        genre_top.append(tag.get_text().strip())
    driver.find_element_by_xpath('//*[@id="TopSellers_btn_next"]').click()
    time.sleep(2)
```

```python
print(topselling)
```

```python
driver.find_element_by_xpath('//*[@id="tab_select_ConcurrentUsers"]').click()
```

```python
beingplay=[]
genre_play=[]
for n in range (9) :
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    ranks = soup.find('div', {'id' : 'ConcurrentUsersTable'}).find_all('div', {'class' : 'tab_item_name'})
    tags = soup.find('div', {'id' : 'ConcurrentUsersTable'}).find_all('span', {'class' : 'top_tag'})
    for rank in ranks :
        beingplay.append(rank.get_text().strip())
    for tag in tags :
        genre_play.append(tag.get_text().strip())
    driver.find_element_by_xpath('//*[@id="ConcurrentUsers_btn_next"]').click()
    time.sleep(2)
```

```python
print(beingplay)
```

```python
import seaborn as sns
from collections import Counter
import numpy as np 
```
